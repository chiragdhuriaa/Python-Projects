```python
def solve_nqueen(board, row, count):
  if row == len(board):
    count[0] += 1
    print_solution(board)
    return

  for col in range(len(board)):
    if is_safe(board, row, col):
      board[row][col] = True
      solve_nqueen(board, row + 1, count)
      board[row][col] = False
```
